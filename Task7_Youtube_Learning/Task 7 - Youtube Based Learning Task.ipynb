```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv("C:\\Users\\prasa\\Downloads\\ds_salaries.csv") 
```

```python
# Check for null values
df.isnull().sum()
```

```python
# Salary distribution
sns.histplot(df['salary_in_usd'], bins=30, kde=True, color="violet")
plt.title('Salary Distribution (USD)')
plt.xlabel('Salary in USD')
plt.ylabel('Frequency')
plt.show()
```

```python
# Average salary by job title
plt.figure(figsize=(12, 6))
df.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False).plot(kind='bar',color="violet")
plt.title('Average Salary by Job Title')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=90)
plt.show()
```

```python
# Count of job types by experience level
sns.countplot(x='experience_level', hue='employment_type', data=df, palette='dark:violet')
plt.title('Job Types vs Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Count')
plt.legend(title='Employment Type')
plt.show()
```

```python
# Remote work impact
sns.boxplot(x='remote_ratio', y='salary_in_usd', data=df, color="violet")
plt.title('Remote Ratio vs Salary')
plt.xlabel('Remote Work %')
plt.ylabel('Salary in USD')
plt.show()
```
