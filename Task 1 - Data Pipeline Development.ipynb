```python
import pandas as pd

df = pd.read_csv("C:\\Users\\prasa\\Downloads\\Titanic.csv")  # import the dataset 
```

```python
df
```

```python
# Checking for missing values
print(df.isnull().sum())
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

le = LabelEncoder() # Encode categorical variables
df['Sex'] = le.fit_transform(df['Sex'])
df['Embarked'] = le.fit_transform(df['Embarked'])
```

```python
df
```

```python
# Separate features and target
X = df.drop(columns=['Survived', 'PassengerId'])
y = df['Survived']
```

```python
# Scale numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
```

```python
# Combine features and target for final output
final_df = X_scaled_df.copy()
final_df['Survived'] = y
```

```python
final_df
```

```python
final_df.to_csv("processed_titanic.csv", index=False)
```
